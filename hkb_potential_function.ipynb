```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
import math
from ipywidgets import interact
```

```python
# Plot the potential function (2.7)
a = 1
b = 1
phis = np.linspace(math.pi,-math.pi,360)
V = -a * np.cos(phis) - b * np.cos(2*phis)
plt.plot(phis,V);
```

```python
# Get a feel for parameters a and b
def f(a,b):
    V = -a * np.cos(phis) - b * np.cos(2*phis)
    plt.plot(phis,V);
interact(f, a=[-10,10,.1], b =[-10,10,.1]);
```

```python
# 2.18
phis = np.linspace(math.pi,-math.pi,360)
r_c = 1
def g(a,r,omega):
    V = -a*(np.cos(phis) + .25 * (((r*omega)**2)/r_c**2)*np.cos(2**phis))
    plt.plot(phis,V);
interact(g, a=[-10,10,.1], r =[-10,10,.1],omega =[0,380,1]);
```
