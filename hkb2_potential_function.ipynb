```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
import math
from ipywidgets import interact
```

## Potential functions.  State will tend to low point.
(See section 2 of the HKB paper)

```python
# Plot the potential function (2.7)
a = 1
b = 1
phis = np.linspace(math.pi,-math.pi,360)
V = -a * np.cos(phis) - b * np.cos(2*phis)
plt.plot(phis,V);
```

```python
# Get a feel for parameters a and b
@interact(a=[0,1,.1], b =[0,1,.1])
def f(a,b):
    V = -a * np.cos(phis) - b * np.cos(2*phis)
    plt.plot(phis,V);
    print(a/(b+.001))
```

```python
# 2.18
phis = np.linspace(math.pi,-math.pi,360)
r_c = 1
@interact(a=[-10,10,.1], r =[-10,10,.1],omega =[0,380,1])
def g(a,r,omega):
    V = -a*(np.cos(phis) + .25 * (((r*omega)**2)/r_c**2)*np.cos(2**phis))
    plt.plot(phis,V);
```
