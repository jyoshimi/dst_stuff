```python
%matplotlib inline 
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np
import math
from ipywidgets import interact
```

```python
times = np.arange(0, 10, .1)
def oscillate(r,phi,omega):
    points = r * np.cos(omega * times + phi)
    plt.plot(times, points);

# r = amplitude; omega = frequency; phi = phase
interact(oscillate, r=[0,10,.1], phi =[0,math.pi,.1], omega = [0,10,.1]);
```

```python
oscillate(1,0,2)
oscillate(1,math.pi,2)
```
