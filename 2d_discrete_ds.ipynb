```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
```

## Define the function to iterate
Feel free to change the value of the constants or  the equation itself

```python
k = .04
E = .1

def f(state):
        
    # Not too hard to generalize to n dimensions
    x = state[0]
    y = state[1]
    
    # Define discrete system here.  This example  based on Dale and Duran, 2013
    # Be aware of coupling terms.  x in y or y in x
    x = x + (-k + x - x**3) + E
    y = k * x - (x**2/2 + x**4/4)
    
    return np.array([x,y])
```

### Set up an initial condition and how long you want to run the system (how many times to iterate the function)

Adjust initial conditions and number of iterations

```python
s = np.array([.1,.4]) # initial condition.  
orbit = [s]
iterations = 100
for t in range(iterations):
    s = f(s)
    orbit = np.vstack([orbit,s])
```

## Time Series Plot

```python
plt.plot(range(len(orbit)), orbit[:,0],'.-', alpha = .5, ms = 15); # Plot the orbit
plt.ylabel("x");
plt.xlabel("iterations");
plt.plot(0, orbit[0,0],'.', ms = 20, c='Red'); # Plot ic as a red dot

# plt.subplot(1, 2, 2)
# plt.scatter(np.zeros(len(orbit)), orbit[:])
# plt.ylabel("x");
# plt.xlabel("iterations");
# plt.xticks([])
```

## Plot an orbit in state space

```python
plt.plot(orbit[:,0], orbit[:,1],'.-', ms = 15, alpha=.2); # Plot the orbit
plt.plot(orbit[0,0], orbit[0,1],'.', ms = 20, c='Red'); # Plot ic
```
