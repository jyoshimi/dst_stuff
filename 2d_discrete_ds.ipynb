```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Wrote this quickly; there may be a better way...
# Easily generalize to n-dimensions
def f(state):
    
    x = state[0]
    y = state[1]
    
    # Define discrete system here.  Ex. roughly based on Dale and Duran, 2013
    k = .01
    E = 0
    x = x + (-k + x - x**3) + E
    y = k * x - (x**2/2 + x**4/4)
    
    # TODO: Efficient?
    return np.array([x,y])
```

```python
s = np.array([1,1]) # initial condition.  add as many components as you want
orbit = [s]
iterations = 100
for t in range(iterations):
    s = f(s)
    orbit = np.vstack([orbit,s])
```

```python
plt.plot(range(len(orbit)), orbit[:,0],'.-', ms = 15); # Plot the orbit
plt.ylabel("x");
plt.xlabel("iterations");

# plt.subplot(1, 2, 2)
# plt.scatter(np.zeros(len(orbit)), orbit[:])
# plt.ylabel("x");
# plt.xlabel("iterations");
# plt.xticks([])
```
