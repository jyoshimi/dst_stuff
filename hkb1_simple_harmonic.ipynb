## Review of harmonic oscillator

(See section 2 of the HKB paper)

r = amplitude

omega = frequency

phi = phase

```python
times = np.arange(0, 10, .1)
@interact(r=[0,10,.1], phi =[0,math.pi,.1], omega = [0,10,.1])
def oscillate(r,phi,omega):
    points = r * np.cos(omega * times + phi)
    plt.plot(times, points);
```

```python
oscillate(1,0,2)
oscillate(1,math.pi,2)
```
