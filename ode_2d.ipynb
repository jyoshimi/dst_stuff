```python
%matplotlib inline 
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np
import itertools
```

## The ODE

```python
# 2d Linear system
def ode(state,t):
    x = state[0]
    y = state[1]
     
    # Vanderpol as an example
    mu = 2
    
    # Main equation here.  Be aware of couplings.
    dx = y
    dy = mu * (1 - x**2)*y-x
    
    # Return the state derivatives
    return [dx,dy]
```

## Plot a few orbits

```python
fig = plt.gcf()
fig.set_size_inches(7,7)

t = np.arange(0, 10, .05) # Times

ic = [1,1] # initial condition 1
orbit = odeint(ode, ic, t) 
plt.plot(orbit[:,0],orbit[:,1], '.-',color='b',ms=5,alpha = .2) 
plt.plot(ic[0],ic[1], '.', color = 'b', ms=10, alpha=.9)

# ic2 = [2,-1] # initial condition 2
# orbit = odeint(ode, ic2, t) 
# plt.plot(orbit[:,0],orbit[:,1], '.-',color='b',ms=5,alpha = .2) 
# plt.plot(ic2[0],ic2[1], '.', color = 'b', ms=10, alpha=.9)
```

## Plot a set of orbits from a grid of initial conditions

```python
fig = plt.gcf()
fig.set_size_inches(7,7)
# A grid of initial conditions to test
t = np.arange(0, 10, .1) # Times
for x,y in itertools.product(range(-4,4), range(-2,2)):
    ic = [x,y] # initial condition
    orbit = odeint(ode, ic, t) # Run the dst; integrate the ode
    plt.plot(orbit[:,0],orbit[:,1], '.-',color='b',ms=5,alpha = .2) 
    plt.plot(ic[0],ic[1], '.', color = 'b', ms=5, alpha=.9) # Initial Condition
```
