```python
%matplotlib inline 
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np
import itertools
```

```python
# 2d Linear system
def ode(state,t):
    x = state[0]
    y = state[1]
     
    # Vanderpol as an example
    mu = 2
    
    # Main equation here
    dx = y
    dy = mu * (1 - x**2)*y-x
    
    # Return the state derivatives
    return [dx,dy]
```

```python
t = np.arange(0, 1, .01) # Times

# A grid of initial conditions to test
for x,y in itertools.product(range(-5,5), range(-5,5)):
    ic = [x,y] # initial condition
    orbit = odeint(ode, ic, t) # Run the dst; integrate the ode
    plt.plot(ic[0],ic[1], '.', color = 'k')
    plt.plot(orbit[:,0],orbit[:,1], color='blue') #Arrows?
```
