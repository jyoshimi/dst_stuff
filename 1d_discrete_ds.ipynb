```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import gridspec
from ipywidgets import interact
```

## Define the function to iterate
Feel free to change the value of r or even the equation itself (though the material below assumes this function, which is the https://en.wikipedia.org/wiki/Logistic_map

```python
def f(x):
    r = 3.9 
    # r = 2.6
    x = r*x*(1-x)
    # x = x*2
    return x
```

### Set up an initial condition and how long you want to run the system (how many times to iterate the function)

You can adjust these things too!

```python
x = .5 # initial condition
orbit = [x]
times = range(50) # Set length of orbit here
for t in times: 
    x = f(x)
    orbit.append(x)
# print(orbit)
```

### Show a Time Series Diagram and a Phase Portrait

```python
gs = gridspec.GridSpec(1, 2, width_ratios=[5, 1]) 

plt.subplot(gs[0])

plt.plot(range(len(orbit)), orbit[:],'.-', ms =10, alpha=.2); # Plot the orbit
plt.ylabel("x");
plt.xlabel("iterations");
plt.plot(0, orbit[0],'.', ms = 20, c='Red'); # Plot ic

plt.subplot(gs[1])
plt.plot(np.zeros(len(orbit)), orbit[:],'.-', ms =10, alpha=.2)
plt.ylabel("x");
plt.xlabel("iterations");
plt.xticks([])
plt.plot(0, orbit[0],'.', ms = 20, c='Red'); # Plot ic
```

# Interaction demo

As you move R compare the number of fixed points you get with  the diagram below, which shows number of fixed points for each value of R

```python
times = range(100) # Set length of orbit here

@interact(r=[1,5,.001]) 
def plot_f(r):
    x = .6 
    plt.plot(0, x,'.', ms = 20, c='Red'); # Plot ic
    for i in times:
        x = r*x*(1-x)
        plt.plot(i,x, '.-', c='b',ms =10, alpha=.2);
```
