```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    r = 3.9
    x = r*x*(1-x)
    #x = x*2
    return x
```

```python
x = .5 # initial condition
orbit = [x]
for t in range(120):
    x = f(x)
    orbit.append(x)
# print(orbit)
```

```python
plt.subplot(1, 2, 1)
plt.plot(range(len(orbit)), orbit[:]); # Plot the orbit
plt.ylabel("x");
plt.xlabel("iterations");

plt.subplot(1, 2, 2)
plt.scatter(np.zeros(len(orbit)), orbit[:])
plt.ylabel("x");
plt.xlabel("iterations");
plt.xticks([])
```

# Interaction demo

```python
from ipywidgets import interact
```

```python
def plot_f(r,x):
    for i in range(100):
        x = r*x*(1-x)
        plt.plot(i,x, '.k');
        
interact(plot_f, r=[1,5,.001], x=[.1,1,.001]) # x is initial condition
```
