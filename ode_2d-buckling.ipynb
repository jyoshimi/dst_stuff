```python
%matplotlib inline 
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np
import itertools
```

```python
# 2d Linear system
def ode(state, t):
    x = state[0]
    y = state[1]
     
    a = 1
    b = 0
    c = .1
    m = 10
    
    # Main equation here
    dx = y
    dy = (1 - m) * (a*x**3 + b + c*y)
    
    # Return the state derivatives
    return [dx,dy]
```

```python
t = np.arange(0, 50, .01) # Times

# A grid of initial conditions to test
for x,y in itertools.product(range(-2,2), range(-2,2)):
    ic = [x,y] # initial condition
    percent = 0 # Percent to truncate
    trunc = int(np.floor(t.size*percent))
    orbit = odeint(ode, ic, t) # Run the dst; integrate the ode
    plt.plot(ic[0],ic[1], '.', color = 'k')
    plt.plot(orbit[trunc:,0],orbit[trunc:,1], color='blue') #Arrows?
```
